```python
FEATURE_COLS = ['danceability', 'energy', 'valence', 'tempo',
    'acousticness', 'speechiness', 'instrumentalness',
    'liveness', 'loudness'
]
```

```python
data = data.dropna()
```

```python

from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import numpy as np
import umap


X = data[FEATURE_COLS].values

X_train, X_test = train_test_split(X, test_size=0.2, random_state=42)

print("Train size:", X_train.shape)
print("Test size:", X_test.shape)


scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)


umap_model = umap.UMAP(
    n_neighbors=30,
    n_components=10,
    min_dist=0.1,
    random_state=42
)

X_train_umap = umap_model.fit_transform(X_train_scaled).astype(np.float32)
X_test_umap  = umap_model.transform(X_test_scaled).astype(np.float32)


k = 7
kmeans = KMeans(n_clusters=k, random_state=42)
train_labels = kmeans.fit_predict(X_train_umap)

sil_train = silhouette_score(X_train_umap, train_labels)
test_labels = kmeans.predict(X_test_umap)
sil_test = silhouette_score(X_test_umap, test_labels)

print("\nSilhouette TRAIN:", sil_train)
print("Silhouette TEST:", sil_test)


X_all = data[FEATURE_COLS].values
X_all_scaled = scaler.transform(X_all)
X_all_umap = umap_model.transform(X_all_scaled).astype(np.float32)

data["tipo_clusters"] = kmeans.predict(X_all_umap)

for i in range(10):
    data[f"umap_{i}"] = X_all_umap[:,i]

print(data["tipo_clusters"].value_counts())
```

```python
import numpy as np
import joblib


kmeans = joblib.load("kmeans_umap.pkl")
scaler = joblib.load("scaler.pkl")
mood_mapping = joblib.load("mood_mapping.pkl")


FEATURE_COLS = [
    'danceability', 'energy', 'valence', 'tempo',
    'acousticness', 'speechiness', 'instrumentalness',
    'liveness', 'loudness'
]


def preprocesar_features(features_dict):

    x = np.array([[features_dict[col] for col in FEATURE_COLS]], dtype=float)
    x_scaled = scaler.transform(x)
    x_umap = umap_model.transform(x_scaled)
    x_umap = x_umap.astype(kmeans.cluster_centers_.dtype)

    return x_umap

```

```python
def obtener_paleta(mood, paletas, paletas_justif):
    colores = paletas[paletas["mood"] == mood]
    justificacion = paletas_justif[
        paletas_justif["mood"] == mood
    ]["justificacion"].iloc[0]

    return {
        "base": colores[colores["tipo"] == "base"]["color_nombre"].tolist(),
        "accent": colores[colores["tipo"] == "accent"]["color_nombre"].tolist(),
        "neutral": colores[colores["tipo"] == "neutral"]["color_nombre"].tolist(),
        "justificacion": justificacion
    }
```

```python
def cluster_a_mood(cluster, mood_map):
    fila = mood_map[mood_map["cluster"] == cluster]
    if fila.empty:
        raise ValueError(f"Cluster {cluster} no tiene mood asignado")
    return fila["mood"].iloc[0]
```

```python
def predecir_mood(features_dict):
    X_scaled = preprocesar_features(features_dict)
    cluster = int(kmeans.predict(X_scaled)[0])
    return cluster
```

```python

def predecir_mood_por_titulo(df, titulo, artista=None):
    titulo = titulo.lower().strip()
    if artista:
        artista = artista.lower().strip()

    df_filtrado = df[df["track_name"].str.lower().str.strip() == titulo]

    if df_filtrado.empty:
        return {"error": f"No se encontró ninguna canción con título '{titulo}'."}

    if artista:
        df_filtrado = df_filtrado[
            df_filtrado["track_artist"].str.lower().str.strip() == artista
        ]
        if df_filtrado.empty:
            return {
                "error": f"Se encontró el título '{titulo}', pero no coincide ningún artista '{artista}'."
            }

    if len(df_filtrado) > 1:
        print("⚠️ Atención: hay varias canciones con ese título/artista. Se usará la primera coincidencia.")
        print(df_filtrado[["track_name", "track_artist"]].head())

    fila = df_filtrado.iloc[0]

    features = fila[FEATURE_COLS].to_dict()

    cluster = predecir_mood(features)

    mood = cluster_a_mood(cluster, mood_map)

    return {
        "title": fila["track_name"],
        "artist": fila["track_artist"],
        "cluster": cluster,
        "mood": mood
    }
```

```python
def generar_outfit(mood, estilo=None, estacion=None, clima=None):
    resultado = {
        "prendas": [],
        "accesorios": [],
        "justificaciones": []
    }

    base_df = outfit_base[outfit_base["mood"] == mood]
    if base_df.empty:
        raise ValueError(f"No existe outfit_base para el mood '{mood}'")

    base = base_df.iloc[0]
    resultado["prendas"].extend(base["prendas_base"].split(";"))
    resultado["accesorios"].extend(base["accesorios_base"].split(";"))
    resultado["justificaciones"].append(base["justificacion"])

    # ---------- Variación por estilo (OPCIONAL) ----------
    if estilo:
        df_estilo = variaciones_estilo[
            (variaciones_estilo["mood"] == mood) &
            (variaciones_estilo["estilo"] == estilo)
        ]
        if not df_estilo.empty:
            fila = df_estilo.iloc[0]
            if pd.notna(fila["prendas_extra"]):
                resultado["prendas"].extend(fila["prendas_extra"].split(";"))
            if pd.notna(fila["accesorios_extra"]):
                resultado["accesorios"].extend(fila["accesorios_extra"].split(";"))
            resultado["justificaciones"].append(fila["justificacion"])

    # ---------- Variación por estación (OPCIONAL) ----------
    if estacion:
        df_estacion = variaciones_contexto[
            (variaciones_contexto["mood"] == mood) &
            (variaciones_contexto["tipo_contexto"] == "estacion") &
            (variaciones_contexto["valor"] == estacion)
        ]
        if not df_estacion.empty:
            fila = df_estacion.iloc[0]
            if pd.notna(fila["prendas_extra"]):
                resultado["prendas"].extend(fila["prendas_extra"].split(";"))
            if pd.notna(fila["accesorios_extra"]):
                resultado["accesorios"].extend(fila["accesorios_extra"].split(";"))
            resultado["justificaciones"].append(fila["justificacion"])

    # ---------- Variación por clima (OPCIONAL) ----------
    if clima:
        df_clima = variaciones_contexto[
            (variaciones_contexto["mood"] == mood) &
            (variaciones_contexto["tipo_contexto"] == "clima") &
            (variaciones_contexto["valor"] == clima)
        ]
        if not df_clima.empty:
            fila = df_clima.iloc[0]
            if pd.notna(fila["prendas_extra"]):
                resultado["prendas"].extend(fila["prendas_extra"].split(";"))
            if pd.notna(fila["accesorios_extra"]):
                resultado["accesorios"].extend(fila["accesorios_extra"].split(";"))
            resultado["justificaciones"].append(fila["justificacion"])

    return resultado
```

```python
def predecir_mood_por_titulo(
    df,
    titulo,
    artista=None,
    estilo=None,
    estacion=None,
    clima=None,
    mood_map=None
):
    titulo = titulo.lower().strip()
    if artista:
        artista = artista.lower().strip()

    df_filtrado = df[df["track_name"].str.lower().str.strip() == titulo]

    if df_filtrado.empty:
        return {"error": f"No se encontró ninguna canción con título '{titulo}'."}

    if artista:
        df_filtrado = df_filtrado[
            df_filtrado["track_artist"].str.lower().str.strip() == artista
        ]
        if df_filtrado.empty:
            return {
                "error": f"Se encontró el título '{titulo}', pero no coincide ningún artista '{artista}'."
            }

    fila = df_filtrado.iloc[0]

    # ---------- Features ----------
    features = fila[FEATURE_COLS].to_dict()

    # ---------- ML ----------
    X_scaled = preprocesar_features(features)
    cluster = int(kmeans.predict(X_scaled)[0])

    # ---------- Cluster → Mood ----------
    mood = cluster_a_mood(cluster, mood_map)

    # ---------- Outfit ----------
    outfit = generar_outfit(
        mood=mood,
        estilo=estilo,
        estacion=estacion,
        clima=clima
    )

    # ---------- Paleta ----------
    paleta = obtener_paleta(mood, paletas, paletas_justif)

    # ---------- Resultado final ----------
    return {
        "title": fila["track_name"],
        "artist": fila["track_artist"],
        "cluster": cluster,
        "mood": mood,
        "outfit": outfit,
        "paleta": paleta
    }
```

### RECOMENDADOR DE CANCIONES SIMILARES


```python
import numpy as np

def distancia_umap(vec1, vec2):
    return np.linalg.norm(vec1 - vec2)
```

```python
def recomendar_por_cancion(df, titulo, artista, top_n=10):

    fila = df[
        (df["track_name"].str.lower() == titulo.lower()) &
        (df["track_artist"].str.lower() == artista.lower())
    ]
    
    if fila.empty:
        return {"error": "Canción no encontrada"}
    
    fila = fila.iloc[0]
    
    base_vec = fila[[f"umap_{i}" for i in range(10)]].values.astype(float)

    df["distancia"] = df.apply(
        lambda row: distancia_umap(
            base_vec,
            row[[f"umap_{i}" for i in range(10)]].values.astype(float)
        ), 
        axis=1
    )

    df_filtrado = df.drop(fila.name)

    recomendadas = df_filtrado.sort_values("distancia").head(top_n)

    return recomendadas[[
        "track_name",
        "track_artist",
        "tipo_clusters",
        "distancia"
    ]]
```
